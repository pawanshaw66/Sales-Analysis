```python
import glob

# List all CSV files in the directory
file_paths = glob.glob('Sales_*.csv')  # Assuming your files are named like 'monthly_sales_January.csv', 'monthly_sales_February.csv', etc.

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each file and load it into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
sales_year = pd.concat(dfs, ignore_index=True)

# Save the merged DataFrame to a new file
sales_year.to_csv('merged_monthly_sales.csv', index=False)
```

# What products are most often sold together?

```python
sales_year.head(2)
```

```python
df = sales_year[sales_year['Order ID'].duplicated(keep=False)]
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
```

```python
df.head()
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import  combinations
from collections import Counter
count = Counter()

for row in df['Grouped']:
  row_list = row.split(',')
  count.update(Counter(combinations(row_list, 2)))
print(count)
```

```python
count.most_common(10)
```
